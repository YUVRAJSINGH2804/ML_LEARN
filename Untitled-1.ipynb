```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import sklearn

```

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')

```

```python
tips.head()
```

```python
#settting A theme for the plot
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load the iris datasset
iris=sns.load_dataset('iris')
iris.head()
```

```python
#scatterplot
sns.scatterplot(x='sepal_length',y= 'petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y= 'petal_width',hue='species',data=iris)
sns.relplot(x='sepal_length',y= 'petal_width',hue='species',data=iris)
```
